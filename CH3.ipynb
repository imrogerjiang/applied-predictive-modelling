```python
import pandas as pd
import numpy as np
import scipy.cluster.hierarchy as sch
import sklearn as skl
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
def cluster_corr(corr_array, inplace=False):
    """
    Rearranges the correlation matrix, corr_array, so that groups of highly 
    correlated variables are next to eachother 
    
    Parameters
    ----------
    corr_array : pandas.DataFrame or numpy.ndarray
        a NxN correlation matrix 
        
    Returns
    -------
    pandas.DataFrame or numpy.ndarray
        a NxN correlation matrix with the columns and rows rearranged
    """
    pairwise_distances = sch.distance.pdist(corr_array)
    linkage = sch.linkage(pairwise_distances, method='complete')
    cluster_distance_threshold = pairwise_distances.max()/2
    idx_to_cluster_array = sch.fcluster(linkage, cluster_distance_threshold, 
                                        criterion='distance')
    idx = np.argsort(idx_to_cluster_array)
    
    if not inplace:
        corr_array = corr_array.copy()
    
    if isinstance(corr_array, pd.DataFrame):
        return corr_array.iloc[idx, :].T.iloc[idx, :]
    return corr_array[idx, :][:, idx]
```

```python
corr_matrix = cluster_corr(bbb.drop(columns=["id", "outcome"]).corr())
plt.figure(figsize=(30,30))
plt.matshow(corr_matrix, fignum=1)
plt.xticks(ticks=range(len(bbb.columns)-2), labels=corr_matrix.columns)
plt.yticks(ticks=range(len(bbb.columns)-2), labels=corr_matrix.columns)
plt.colorbar()
plt.show()
```

```python
drop_corr = corr_matrix
for c in drop_corr:
    drop_corr[c].loc[c] = 0    
```

```python
max_corr=drop_corr.max().max()
while max_corr > 0.75:
    max1 = drop_corr.max()
    max1_idx = drop_corr.idxmax()
    max2_idx = max1.idxmax()
    var1, var2=max1_idx[max2_idx], max2_idx

    if drop_corr[var1].sum()>drop_corr[var2].sum():
        drop_corr = drop_corr.drop(columns=var1, index=var1)
    else:
        drop_corr = drop_corr.drop(columns=var2, index=var2)
        
    max_corr=drop_corr.max().max()
```

```python
plt.figure(figsize=(30,30))
plt.matshow(drop_corr, fignum=1)
plt.xticks(ticks=range(len(drop_corr)), labels=drop_corr.columns)
plt.yticks(ticks=range(len(drop_corr)), labels=drop_corr.columns)
plt.colorbar()
plt.show()
```
